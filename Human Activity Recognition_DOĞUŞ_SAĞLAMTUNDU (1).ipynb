```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
X = datas.iloc[:,:561]
y = datas.iloc[:, 562:563]
```

```python
scaler = StandardScaler()
X= scaler.fit_transform(X) #std sonrası array hali
X = pd.DataFrame(X)
X                               #std sonrası arrayin dataframe hali
```

```python
from sklearn import decomposition
```

```python
PCA(copy=True, iterated_power='auto', n_components=0.9, random_state=None,
  svd_solver='auto', tol=0.0, whiten=False)
```

```python
pca = decomposition.PCA(n_components=0.9)
```

```python
pca_data = pca.fit_transform(X)
```

```python
print(pca.explained_variance_ratio_.sum())
```

```python
pca_data = pd.DataFrame(pca_data)
pca_data
```

```python
X = pca_data
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.model_selection import KFold, cross_val_score
```

```python
#Finding optimal K value
acc = []
# Will take some time
from sklearn import metrics
for i in range(1,10):
    neigh = KNN(n_neighbors = i).fit(X_train,y_train)
    yhat = neigh.predict(X_test)
    acc.append(metrics.accuracy_score(y_test, yhat))
    
plt.figure(figsize=(10,6))
plt.plot(range(1,10),acc,color = 'blue',linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
plt.title('accuracy vs. K Value')
plt.xlabel('K')
plt.ylabel('Accuracy')
print("Maximum accuracy:-",max(acc),"at K =",acc.index(max(acc)))
```

```python
model = KNN(algorithm= 'auto', n_neighbors= 8, p= 1, weights= 'distance')

model.fit(X_train, y_train)

#eğitim seti üzerinde tahminler yapma ve f1 hesaplama işlemi
y_pred = model.predict(X_test)
f_1 = f1_score(y_pred, y_test, average='macro')
```

```python
print("F1 score is: ",f_1)
print("Model of accuracy is: ", accuracy_score(y_test, y_pred))
```

```python
cros_val= cross_val_score(model, X_train, y_train, cv=10)
cros_val
```

```python
sum= 0
for i in range(len(cros_val)):
    sum = sum + cros_val[i]
    
print("Average of accuracy is: ",sum/10)
```

```python
cf_matrix= confusion_matrix(y_test, y_pred)
cf = pd.DataFrame(cf_matrix, columns=np.unique(y_test), index = np.unique(y_test))
cf.index.name = 'Actual Values'
cf.columns.name = 'Predicted Values'
```
